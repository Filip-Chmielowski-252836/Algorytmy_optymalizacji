```python
import matplotlib.pyplot as plt

# ustawiamy fizyczny rozmiar wykresu
plt.figure(figsize=(3,3))
# definiujemy prostokąty (lewy dolny róg + wysokość i szerokość)
rectangle1=plt.Rectangle(xy=(20,30), width=40, height=50, edgecolor='blue', facecolor='royalblue', linewidth=1)
rectangle2=plt.Rectangle(xy=(10,0), width=30, height=30, edgecolor='blue', facecolor='royalblue', linewidth=1)
rectangle3=plt.Rectangle(xy=(70,0), width=20, height=90, edgecolor='blue', facecolor='royalblue', linewidth=1)
# zapisujemy uchwyt wyresu
figure = plt.gcf()
# zapisujemy uchwyt aktualych osi (Current Axes) 
axes = figure.gca()
# dodajemy prostokąty do obiektu osi
axes.add_patch(rectangle1)
axes.add_patch(rectangle2)
axes.add_patch(rectangle3)
# ustawiamy skalę wykresu
plt.xlim([0,100])
plt.ylim([0,100])
# i plotujemy wykres
plt.show()
```
